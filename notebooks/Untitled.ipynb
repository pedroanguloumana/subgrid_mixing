```python
%load_ext autoreload
%autoreload 2
import xarray as xr
import numpy as np
from glob import glob
import matplotlib.pyplot as plt
import matplotlib.colors as colors
from src.sam import *
```

```python
Tkoff = load_2d_TKScaleOff()
Tkon = load_2d_TKScale0p2()
```

```python
grouped_Tkoff = Tkoff.PW.groupby('time')
grouped_Tkon = Tkon.PW.groupby('time')
for (_, data_off), (_, data_on) in zip(grouped_Tkoff, grouped_Tkon):
    fig, ax = plt.subplots(ncols=2)
    cmap = plt.cm.jet.copy()
    norm = colors.Normalize(vmin=0, vmax=70)
    data_off.plot(ax=ax[0], cmap=cmap, norm=norm)
    ax[0].set_title('TK Off')
    data_on.plot(ax=ax[1], cmap=cmap, norm=norm)
    ax[1].set_title('TK On')
    fig.tight_layout(pad=1.0)
```

```python
data_off = Tkoff.OM200
data_on = Tkon.OM200

fig, ax = plt.subplots()
bins = np.linspace(-0.5, 0.1, 100)
data_off.plot.hist(ax=ax, bins=bins, histtype='step', density=True, label='TK off')
data_on.plot.hist(ax=ax, bins=bins, histtype='step', density=True, label='TK on')
ax.legend()
```
